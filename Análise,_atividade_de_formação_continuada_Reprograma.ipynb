```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x=most_frequent_positions.index, y=most_frequent_positions, palette='viridis')
plt.title('Top 10 most frequent positions')
plt.xlabel('Position')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
desired_positions = ['Data Scientist', 'Data Engineer', 'Data Analyst']

filtered_df = df[df['job_title'].isin(desired_positions)]

position_counts = filtered_df['job_title'].value_counts()
print(position_counts)

grouped_data = filtered_df.groupby(['work_year', 'job_title'])['salary_in_usd'].mean().unstack()

plt.figure(figsize=(8, 6))
sns.lineplot(data=grouped_data, markers=True, dashes=False)

plt.xlabel('Year')
plt.ylabel('Salary in USD')
plt.title('Salary trends in top 3 positions over the years')

plt.xticks(grouped_data.index)

plt.legend(title='Positions', bbox_to_anchor=(1, 1))
plt.show()
```

```python
selected_jobs = ['Data Scientist', 'Data Engineer', 'Data Analyst']
df_filtered = df[df['job_title'].isin(selected_jobs)]

plt.figure(figsize=(8, 6))
sns.barplot(x='employee_residence', y='salary_in_usd', data=df_filtered,
            order=df_filtered.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending=False).index)
plt.xlabel('Geographical location')
plt.ylabel('Salary in USD')
plt.title('USD salary by geographical location for Data Scientist, Data Engineer, and Data Analyst')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Supondo que você já tenha carregado seu DataFrame df

selected_jobs = ['Data Scientist', 'Data Engineer', 'Data Analyst']
df_filtered = df[df['job_title'].isin(selected_jobs)]

# Calcula a média do salary_in_usd agrupado por employee_residence
mean_salary_by_location = df_filtered.groupby('employee_residence')['salary_in_usd'].mean().reset_index()

# Ordena o DataFrame pela média do salário em ordem decrescente
mean_salary_by_location = mean_salary_by_location.sort_values(by='salary_in_usd', ascending=False)

# Cria o gráfico de barras
plt.figure(figsize=(8, 6))
sns.barplot(x='employee_residence', y='salary_in_usd', data=mean_salary_by_location,
            order=mean_salary_by_location['employee_residence'])
plt.xlabel('Geographical location')
plt.ylabel('Average Salary in USD')
plt.title('Average USD salary by geographical location for Data Scientist, Data Engineer, and Data Analyst')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='experience_level', y='salary_in_usd', data=df, order=df.groupby('experience_level')['salary_in_usd'].mean().sort_values().index)
plt.xlabel('Experience level')
plt.ylabel('Salary in USD')
plt.title('USD salary by experience level')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='work_year', y='salary_in_usd', hue='experience_level', data=df, ci=None)
plt.xlabel('Year')
plt.ylabel('Average salary in USD')
plt.title('Average salaries by year and experience level')
plt.show()
```

```python

data_scientists = df[df['job_title'] == 'Data Scientist']


location_counts = data_scientists['employee_residence'].value_counts().reset_index()

location_counts.columns = ['employee_residence', 'CountDataScientists']

most_common_location = location_counts.loc[location_counts['CountDataScientists'].idxmax()]

print("Local mais comum para Data Scientists:", most_common_location['employee_residence'])
print("Número de Data Scientists nesse local:", most_common_location['CountDataScientists'])
```
